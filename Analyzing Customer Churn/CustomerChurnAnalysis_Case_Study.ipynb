# 📉 Customer Churn Analysis in Excel

## 🧠 Objective
To explore churn behavior of telecom customers using Excel and uncover insights that can help reduce churn at Databel, a fictitious subscription-based company.

## 📁 Project Overview
- Investigated churn data using Excel
- Used PivotTables, calculated fields, and dashboards
- Identified high-risk demographics and behaviors
- Presented findings with visual dashboards

## 📊 Key Business Questions & Answers

**Q1.** Does our dataset contain any duplicate values?
**A:** No duplicate values exist in the dataset.

**Q2.** What's the total churn rate for Databel?
**A:** 26.86%

**Q3.** Which of the following is part of the top 3 churn reasons?
**A:** d) Competitor had better devices

**Q4.** What % of customers churned due to 'Competitor made better offer'?
**A:** 37.64%

**Q5.** Which demographic group has the highest churn rate?
**A:** Senior citizens — almost 40%

**Q6.** Which age group makes up the lowest number of customers but highest churn rate?
**A:** 79–88

**Q7.** What's the churn rate for people on an unlimited plan who consume < 5 GB?
**A:** 34.69%

**Q8.** Which state with international plans had the highest churn?
**A:** CA

**Q9.** Account Length group with biggest gap between 1-Year and 2-Year churn?
**A:** a) 37–48

**Q10.** Total number of customers Databel has served?
**A:** 6687

**Q11.** Least used reason under competitor churn analysis?
**A:** b) Competitor offered higher download speeds

**Q12.** Churn rate for customers aged 19–28?
**A:** 21.96%

**Q13.** States with 25% churn rate under international plan?
**A:** Wisconsin (WI), Idaho (ID), North Dakota (ND), Rhode Island (RI)

## 📈 Insights Summary
- Senior citizens are most likely to churn
- Major churn reason: competitors offering better deals
- Younger customers churn less
- Customers using low data on unlimited plans churn more
- Targeted retention strategies can focus on contract types and senior users
